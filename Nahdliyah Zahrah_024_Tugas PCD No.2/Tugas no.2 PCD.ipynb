```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("lena.jpg")
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb
plt.imshow(image_rgb)
```

# A. Box kernel

```python
# Kernel
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9
```

```python
lena_smoothed = cv2.filter2D(image_rgb, -1, kernel)
plt.imshow(lena_smoothed)
```

# B. Gaussian *kernel*.

```python
# Kernel
kernel = np.array([[0.3673, 0.6065, 0.3679],
                   [0.6065, 1.0000, 0.6065],
                   [0.3679, 0.6065, 0.3679]]) / 4.8976
```

```python
lena_filtered = cv2.filter2D(image_rgb, -1, kernel)
plt.imshow(lena_filtered)
```

```python
grayscale_img=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
hsv_img=cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

fig,axs= plt.subplots(1,6, figsize=(10,4))

# Kernel
kernel_a = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9

# Kernel
kernel_b = np.array([[0.3673, 0.6065, 0.3679],
                   [0.6065, 1.0000, 0.6065],
                   [0.3679, 0.6065, 0.3679]]) / 4.8976

# Mengaplikasikan filter 2D pada gambar menggunakan kedua kernel
smoothed_lena_a = cv2.filter2D(grayscale_img, -1, kernel_a)
smoothed_lena_b = cv2.filter2D(grayscale_img, -1, kernel_b)

axs[0].imshow(image_rgb)
axs[0].set_title("Original Image Shape")

lena_smoothed = cv2.filter2D(image_rgb, -1, kernel)
axs[1].imshow(lena_smoothed)
axs[1].set_title("Box Kernel")


lena_filtered = cv2.filter2D(image_rgb, -1, kernel)
axs[2].imshow(lena_filtered)
axs[2].set_title("Gaussian Kernel")


axs[3].imshow(grayscale_img,cmap="gray")
axs[3].set_title("Grayscale Image")

lena_smoothed = cv2.filter2D(grayscale_img, -1, kernel)
axs[4].imshow(smoothed_lena_a, cmap='gray')
axs[4].set_title("Box Kernel GS")


lena_filtered = cv2.filter2D(grayscale_img, -1, kernel)
axs[5].imshow(smoothed_lena_b, cmap='gray')
axs[5].set_title("Gaussian Kernel GS")


for ax in axs:
  ax.set_xticks([])
  ax.set_yticks([])

plt.tight_layout()
plt.show()
```

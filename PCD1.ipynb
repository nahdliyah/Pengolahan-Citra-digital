```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("lena.jpg")
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb
```

## Mengganti Warna

```python
grayscale_img=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
hsv_img=cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

fig,axs= plt.subplots(1,4, figsize=(10,4))

axs[0].imshow(image_rgb)
axs[0].set_title("Original Image Shape")

axs[1].imshow(grayscale_img,cmap="gray")
axs[1].set_title("Grayscale Image 1")

axs[3].imshow(grayscale_img)
axs[3].set_title("Grayscale Image 1")

axs[2].imshow(hsv_img)
axs[2].set_title("HSV Image")

for ax in axs:
  ax.set_xticks([])
  ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Citra Negatif

```python
img_negatif=abs(255-image_rgb)
plt.imshow(img_negatif)
```

```python
image = cv2.imread("lena.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
image_negatif = abs(255-image_rgb)
image_negatif2 = abs(255-image_gray)

plt.imshow(image_negatif2)
```

```python
image = cv2.imread("lena.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
image_negatif = cv2.bitwise_not(image_rgb) 
image_negatif2 = cv2.bitwise_not(image_rgb)

plt.imshow(image_negatif2)
```

## Transformasi Power-Law (Gamma)

```python
gamma_corrected = np.array (255*(image_rgb/255)**0, dtype="uint8")
plt.imshow(gamma_corrected)
plt.show
```

```python
img = cv2.imread('lena.jpg')
c = 255/(np.log(1 + np.max(img)))
log_transformed = c * np.log(1 + img)
log_transformed = np.array(log_transformed, dtype = np.uint8)
log_transformed
```

## Histogram

```python
img=cv2.imread('lena.jpg',0)

#find frequency of pixels in range 0-255
histr=cv2.calcHist([img],[0],None,[256],[0,256])

plt.plot(histr)
plt.show
```

```python


#alternative way to find hostogram of an image
plt.hist(img.ravel(),256,[0,256])
plt.show
```

* Ekualisasi Histogram

```python
equ=cv2.equalizeHist(img)
plt.hist(equ.ravel(),256,[0,256])
plt.show
```

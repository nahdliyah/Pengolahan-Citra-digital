# NOISE 

### GAUSSIAN NOISE

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Baca citra dan konversi ke grayscale
img = cv2.imread('ikan.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Buat array untuk menyimpan Gaussian noise
gauss_noise = np.zeros_like(img_gray, dtype=np.uint8)
cv2.randn(gauss_noise, 128, 20)  # Membuat Gaussian noise dengan mean=128 dan std=20
gauss_noise = (gauss_noise * 0.5).astype(np.uint8)  # Skala nilai noise

# Tambahkan Gaussian noise ke citra grayscale
gn_img = cv2.add(img_gray, gauss_noise)

# Plot citra asli, Gaussian noise, dan citra dengan noise
fig = plt.figure(figsize=(12, 4), dpi=100)

# Plot citra asli
fig.add_subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.axis("off")
plt.title("Original")

# Plot Gaussian noise
fig.add_subplot(1, 3, 2)
plt.imshow(gauss_noise, cmap='gray')
plt.axis("off")
plt.title("Gaussian Noise")

# Plot citra dengan Gaussian noise
fig.add_subplot(1, 3, 3)
plt.imshow(gn_img, cmap='gray')
plt.axis("off")
plt.title("Combined")

plt.tight_layout()
plt.show()
```

### Rayleigh Noise 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def add_rayleigh_noise(image, scale=50):
    """
    Menambahkan Rayleigh Noise ke citra.
    
    Parameters:
        image (numpy array): Citra sebagai numpy array.
        scale (float): Skala faktor Rayleigh noise.
        
    Returns:
        numpy array: Citra dengan Rayleigh Noise.
    """
    noisy_image = np.copy(image)
    height, width = image.shape
    
    # Generate Rayleigh noise
    noise = np.random.rayleigh(scale, (height, width))
    
    # Add noise to the image
    noisy_image = np.clip(noisy_image + noise, 0, 255)
    noisy_image = noisy_image.astype(np.uint8)
    
    return noisy_image

# Baca citra dan konversi ke grayscale
img = cv2.imread('ikan.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tambahkan Rayleigh Noise
rayleigh_noise = add_rayleigh_noise(img_gray, scale=50)

# Plot citra asli dan citra dengan Rayleigh Noise
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rayleigh_noise, cmap='gray')
plt.title('Rayleigh Noise')
plt.axis('off')

plt.show()
```

### Exponential Noise

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def add_exponential_noise(image, scale=0.1):
    """
    Menambahkan Exponential Noise ke citra grayscale.
    
    Parameters:
        image (numpy array): Citra grayscale sebagai numpy array.
        scale (float): Parameter skala distribusi eksponensial.
        
    Returns:
        numpy array: Citra dengan Exponential Noise.
    """
    # Menghasilkan noise dengan distribusi eksponensial
    noise = np.random.exponential(scale, image.shape)
    
    # Menambahkan noise ke citra
    noisy_image = image + noise
    
    # Memastikan nilai piksel tetap dalam rentang 0-255
    noisy_image = np.clip(noisy_image, 0, 255)
    
    # Konversi tipe data float ke uint8 (sama dengan tipe data citra)
    noisy_image = noisy_image.astype(np.uint8)
    
    return noisy_image

# Baca citra sebagai grayscale
img = cv2.imread('ikan.jpg', cv2.IMREAD_GRAYSCALE)

# Periksa apakah citra berhasil dibaca
if img is None:
    print("Error: Gambar tidak dapat dibaca.")
else:
    # Tambahkan Exponential Noise dengan skala 50
    exponential_noise = add_exponential_noise(img, scale=50)
    
    # Plot citra asli dan citra dengan Exponential Noise
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Citra Asli')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(exponential_noise, cmap='gray')
    plt.title('Citra dengan Exponential Noise')
    plt.axis('off')
    
    plt.show()
```

### Uniform Noise

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def add_uniform_noise(image, intensity_range=(0, 50)):
    """
    Menambahkan Uniform Noise ke citra grayscale.
    
    Parameters:
        image (numpy array): Citra grayscale sebagai numpy array.
        intensity_range (tuple): Rentang nilai noise yang akan ditambahkan.
        
    Returns:
        numpy array: Citra dengan Uniform Noise.
    """
    # Mendapatkan ukuran citra
    height, width = image.shape
    
    # Mendefinisikan rentang nilai noise
    min_intensity, max_intensity = intensity_range
    
    # Membuat noise dengan distribusi seragam
    noise = np.random.randint(min_intensity, max_intensity + 1, size=(height, width))
    
    # Menambahkan noise ke citra
    noisy_image = image + noise
    
    # Memastikan nilai piksel tetap dalam rentang 0-255
    noisy_image = np.clip(noisy_image, 0, 255)
    
    # Konversi tipe data float ke uint8 (sama dengan tipe data citra)
    noisy_image = noisy_image.astype(np.uint8)
    
    return noisy_image

# Baca citra sebagai grayscale
img = cv2.imread('ikan.jpg', cv2.IMREAD_GRAYSCALE)

# Periksa apakah citra berhasil dibaca
if img is None:
    print("Error: Gambar tidak dapat dibaca.")
else:
    # Tambahkan Uniform Noise dengan rentang nilai (0, 50)
    uniform_noise = add_uniform_noise(img, intensity_range=(0, 50))
    
    # Plot citra asli dan citra dengan Uniform Noise
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Citra Asli')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(uniform_noise, cmap='gray')
    plt.title('Citra dengan Uniform Noise')
    plt.axis('off')
    
    plt.show()
```

### Salt-and-Pepper Noise

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def add_salt_and_pepper_noise(image, amount=0.05):
    """
    Menambahkan Salt and Pepper Noise ke citra.
    
    Parameters:
        image (numpy array): Citra sebagai numpy array.
        amount (float): Persentase pixel yang akan diganggu.
        
    Returns:
        numpy array: Citra dengan Salt and Pepper Noise.
    """
    noisy_image = np.copy(image)
    height, width = image.shape
    
    # Hitung jumlah pixel yang akan diganggu
    num_pixels = int(amount * height * width)
    
    # Pilih secara acak pixel untuk diganggu
    for _ in range(num_pixels):
        # Pilih koordinat pixel secara acak
        x = np.random.randint(0, width)
        y = np.random.randint(0, height)
        
        # Tentukan nilai noise (hitam atau putih)
        if np.random.rand() < 0.5:
            noisy_image[y, x] = 0  # Salt (nilai hitam)
        else:
            noisy_image[y, x] = 255  # Pepper (nilai putih)
    
    return noisy_image

# Baca citra dan konversi ke grayscale
img = cv2.imread('ikan.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tambahkan Salt and Pepper Noise
salt_paper_noise = add_salt_and_pepper_noise(img_gray, amount=0.05)  # Persentase noise 5%

# Plot citra asli dan citra dengan Salt and Pepper Noise
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(salt_paper_noise, cmap='gray')
plt.title('Salt and Pepper Noise')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def median_filter(img, kernel_size):
    img_height, img_width = img.shape
    pad_size = kernel_size // 2
    padded_img = np.pad(img, pad_size, mode='constant')
    filtered_img = np.zeros_like(img)

    for i in range(img_height):
        for j in range(img_width):
            neighborhood = padded_img[i:i+kernel_size, j:j+kernel_size].flatten()
            median_value = np.median(neighborhood)
            filtered_img[i, j] = median_value

    return filtered_img

def calculate_psnr(original_img, filtered_img):
    mse = np.mean((original_img - filtered_img) ** 2)
    if mse == 0:
        return float('inf')
    max_pixel = 255.0
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr

def calculate_mse(original_img, filtered_img):
    mse = np.mean((original_img - filtered_img) ** 2)
    return mse

# Load original image
original_img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)

# Generate and filter noisy images
kernel_size = 3

gn_median = median_filter(gn_img, kernel_size)
uniform_median = median_filter(uniform_noise, kernel_size)
exponential_median = median_filter(exponential_noise, kernel_size)
salt_paper_median = median_filter(salt_paper_noise, kernel_size)
rayleigh_median = median_filter(rayleigh_noise, kernel_size)

# Calculate PSNR and MSE for each filtered image
psnr_gn = calculate_psnr(gn_img, gn_median)
mse_gn = calculate_mse(gn_img, gn_median)

psnr_uniform = calculate_psnr(uniform_noise, uniform_median)
mse_uniform = calculate_mse(uniform_noise, uniform_median)

psnr_exponential = calculate_psnr(exponential_noise, exponential_median)
mse_exponential = calculate_mse(exponential_noise, exponential_median)

psnr_salt_paper = calculate_psnr(salt_paper_noise, salt_paper_median)
mse_salt_paper = calculate_mse(salt_paper_noise, salt_paper_median)

psnr_rayleigh = calculate_psnr(rayleigh_noise, rayleigh_median)
mse_rayleigh = calculate_mse(rayleigh_noise, rayleigh_median)

# Plot the filtered images with titles
fig, axs = plt.subplots(1, 5, figsize=(15, 4))

axs[0].imshow(gn_median, cmap='gray')
axs[0].set_title(f"Gaussian_Median\nPSNR: {psnr_gn:.2f} dB\nMSE: {mse_gn:.2f}")

axs[1].imshow(rayleigh_median, cmap='gray')
axs[1].set_title(f"Rayleigh_Median\nPSNR: {psnr_rayleigh:.2f} dB\nMSE: {mse_rayleigh:.2f}")

axs[2].imshow(exponential_median, cmap='gray')
axs[2].set_title(f"Exponential_Median\nPSNR: {psnr_exponential:.2f} dB\nMSE: {mse_exponential:.2f}")

axs[3].imshow(uniform_median, cmap='gray')
axs[3].set_title(f"Uniform_Median\nPSNR: {psnr_uniform:.2f} dB\nMSE: {mse_uniform:.2f}")

axs[4].imshow(salt_paper_median, cmap='gray')
axs[4].set_title(f"Salt_Paper_Median\nPSNR: {psnr_salt_paper:.2f} dB\nMSE: {mse_salt_paper:.2f}")

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

### MAX FILTER

# Mean Filters

### ARITHMATIC MEAN FILTER

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def arithmetic_mean_filter(img, kernel_size):
    img_height, img_width = img.shape
    pad_size = kernel_size // 2
    padded_img = np.pad(img, pad_size, mode='constant')
    filtered_img = np.zeros_like(img)

    for i in range(img_height):
        for j in range(img_width):
            # Extract neighborhood
            neighborhood = padded_img[i:i+kernel_size, j:j+kernel_size]
            # Calculate mean of the neighborhood
            mean_value = np.mean(neighborhood)
            # Assign mean value to filtered image
            filtered_img[i, j] = mean_value

    return filtered_img

# Load image
image = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)

# Define kernel size (should be odd)
kernel_size = 3

# Apply max filter
gn_arithmatic = arithmetic_mean_filter(gn_img, kernel_size)
uniform_arithmatic = arithmetic_mean_filter(uniform_noise, kernel_size)
exponential_arithmatic = arithmetic_mean_filter(exponential_noise, kernel_size)
salt_paper_arithmatic = arithmetic_mean_filter(salt_paper_noise, kernel_size)
rayleigh_arithmatic = arithmetic_mean_filter(rayleigh_noise, kernel_size)


# Calculate PSNR and MSE for each filtered image
psnr_gn = calculate_psnr(gn_img, gn_arithmatic)
mse_gn = calculate_mse(gn_img, gn_arithmatic)

psnr_uniform = calculate_psnr(uniform_noise, uniform_arithmatic)
mse_uniform = calculate_mse(uniform_noise, uniform_arithmatic)

psnr_exponential = calculate_psnr(exponential_noise, exponential_arithmatic)
mse_exponential = calculate_mse(exponential_noise, exponential_arithmatic)

psnr_salt_paper = calculate_psnr(salt_paper_noise, salt_paper_arithmatic)
mse_salt_paper = calculate_mse(salt_paper_noise, salt_paper_arithmatic)

psnr_rayleigh = calculate_psnr(rayleigh_noise, rayleigh_arithmatic)
mse_rayleigh = calculate_mse(rayleigh_noise, rayleigh_arithmatic)

# Plot the filtered images with titles
fig, axs = plt.subplots(1, 5, figsize=(15, 4))

axs[0].imshow(gn_arithmatic, cmap='gray')
axs[0].set_title(f"Gaussian_Arithmatic\nPSNR: {psnr_gn:.2f} dB\nMSE: {mse_gn:.2f}")

axs[1].imshow(rayleigh_arithmatic, cmap='gray')
axs[1].set_title(f"Rayleigh_arithmatic\nPSNR: {psnr_rayleigh:.2f} dB\nMSE: {mse_rayleigh:.2f}")

axs[2].imshow(exponential_arithmatic, cmap='gray')
axs[2].set_title(f"Exponential_arithmatic\nPSNR: {psnr_exponential:.2f} dB\nMSE: {mse_exponential:.2f}")

axs[3].imshow(uniform_arithmatic, cmap='gray')
axs[3].set_title(f"Uniform_arithmatic\nPSNR: {psnr_uniform:.2f} dB\nMSE: {mse_uniform:.2f}")

axs[4].imshow(salt_paper_arithmatic, cmap='gray')
axs[4].set_title(f"Salt_paper_arithmatic\nPSNR: {psnr_salt_paper:.2f} dB\nMSE: {mse_salt_paper:.2f}")

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

### GEOMETRIC MEAN FILTER

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def geometric_mean_filter(img, kernel_size):
    img_height, img_width = img.shape
    pad_size = kernel_size // 2
    padded_img = np.pad(img, pad_size, mode='constant')
    filtered_img = np.zeros_like(img, dtype=np.float64)

    for i in range(img_height):
        for j in range(img_width):
            # Extract neighborhood
            neighborhood = padded_img[i:i+kernel_size, j:j+kernel_size]
            # Calculate geometric mean of the neighborhood
            geometric_mean_value = np.exp(np.mean(np.log(neighborhood + 1)))
            # Assign geometric mean value to filtered image
            filtered_img[i, j] = geometric_mean_value

    return filtered_img.astype(np.uint8)

# Load image
image = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)

# Define kernel size (should be odd)
kernel_size = 3


# Apply max filter
gn_geometric = geometric_mean_filter(gn_img, kernel_size)
uniform_geometric = geometric_mean_filter(uniform_noise, kernel_size)
exponential_geometric = geometric_mean_filter(exponential_noise, kernel_size)
salt_paper_geometric = geometric_mean_filter(salt_paper_noise, kernel_size)
rayleigh_geometric = geometric_mean_filter(rayleigh_noise, kernel_size)

# Display original and geometric filtered images
# fig, axs = plt.subplots(1, 5, figsize=(10, 4))

# Calculate PSNR and MSE for each filtered image
psnr_gn = calculate_psnr(gn_img, gn_geometric)
mse_gn = calculate_mse(gn_img, gn_geometric)

psnr_uniform = calculate_psnr(uniform_noise, uniform_geometric)
mse_uniform = calculate_mse(uniform_noise, uniform_geometric)

psnr_exponential = calculate_psnr(exponential_noise, exponential_geometric)
mse_exponential = calculate_mse(exponential_noise, exponential_geometric)

psnr_salt_paper = calculate_psnr(salt_paper_noise, salt_paper_geometric)
mse_salt_paper = calculate_mse(salt_paper_noise, salt_paper_geometric)

psnr_rayleigh = calculate_psnr(rayleigh_noise, rayleigh_geometric)
mse_rayleigh = calculate_mse(rayleigh_noise, rayleigh_geometric)

# Plot the filtered images with titles
fig, axs = plt.subplots(1, 5, figsize=(15, 4))

axs[0].imshow(gn_geometric, cmap='gray')
axs[0].set_title(f"Gaussian_Geometric\nPSNR: {psnr_gn:.2f} dB\nMSE: {mse_gn:.2f}")

axs[1].imshow(rayleigh_geometric, cmap='gray')
axs[1].set_title(f"Rayleigh_geometric\nPSNR: {psnr_rayleigh:.2f} dB\nMSE: {mse_rayleigh:.2f}")

axs[2].imshow(exponential_geometric, cmap='gray')
axs[2].set_title(f"Exponential_geometric\nPSNR: {psnr_exponential:.2f} dB\nMSE: {mse_exponential:.2f}")

axs[3].imshow(uniform_geometric, cmap='gray')
axs[3].set_title(f"Uniform_geometric\nPSNR: {psnr_uniform:.2f} dB\nMSE: {mse_uniform:.2f}")

axs[4].imshow(salt_paper_geometric, cmap='gray')
axs[4].set_title(f"Salt_paper_geometric\nPSNR: {psnr_salt_paper:.2f} dB\nMSE: {mse_salt_paper:.2f}")

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

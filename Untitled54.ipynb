```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar dalam mode grayscale
lena = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)

# Mendefinisikan kernel A dan B
kernel_a = np.array([[1/9, 1/9, 1/9],
                     [1/9, 1/9, 1/9],
                     [1/9, 1/9, 1/9]])

kernel_b = np.array([[1/16, 1/16, 1/16, 1/16],
                     [1/16, 1/16, 1/16, 1/16],
                     [1/16, 1/16, 1/16, 1/16],
                     [1/16, 1/16, 1/16, 1/16]])

# Mengaplikasikan filter 2D pada gambar menggunakan kedua kernel
smoothed_lena_a = cv2.filter2D(lena, -1, kernel_a)
smoothed_lena_b = cv2.filter2D(lena, -1, kernel_b)

# Menampilkan kedua hasil filter
plt.subplot(1, 2, 1)
plt.imshow(smoothed_lena_a, cmap='gray')  # Menggunakan colormap 'gray' untuk menampilkan gambar grayscale
plt.title('Smoothed with Kernel A')

plt.subplot(1, 2, 2)
plt.imshow(smoothed_lena_b, cmap='gray')
plt.title('Smoothed with Kernel B')

plt.show()
```
